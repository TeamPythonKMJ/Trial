```python
# import dependencies
import pandas as pd
```

```python
# Import datasets
C_compounds_df = pd.read_csv('Output/C_compounds_composition.csv', index_col=0)
Cr_compounds_df = pd.read_csv('Output/Cr_compounds_composition.csv', index_col=0)
Fe_compounds_df = pd.read_csv('Output/Fe_compounds_composition.csv', index_col=0)
Hf_compounds_df = pd.read_csv('Output/Hf_compounds_composition.csv', index_col=0)
Mo_compounds_df = pd.read_csv('Output/Mo_compounds_composition.csv', index_col=0)
Nb_compounds_df = pd.read_csv('Output/Nb_compounds_composition.csv', index_col=0)
Ni_compounds_df = pd.read_csv('Output/Ni_compounds_composition.csv', index_col=0)
Ta_compounds_df = pd.read_csv('Output/Ta_compounds_composition.csv', index_col=0)
Ti_compounds_df = pd.read_csv('Output/Ti_compounds_composition.csv', index_col=0)
V_compounds_df = pd.read_csv('Output/V_compounds_composition.csv', index_col=0)
W_compounds_df = pd.read_csv('Output/W_compounds_composition.csv', index_col=0)
Zr_compounds_df = pd.read_csv('Output/Zr_compounds_composition.csv', index_col=0)

C_properties_df = pd.read_csv('Output/C_compounds_properties.csv', index_col=0)
Cr_properties_df = pd.read_csv('Output/Cr_compounds_properties.csv', index_col=0)
Fe_properties_df = pd.read_csv('Output/Fe_compounds_properties.csv', index_col=0)
Hf_properties_df = pd.read_csv('Output/Hf_compounds_properties.csv', index_col=0)
Mo_properties_df = pd.read_csv('Output/Mo_compounds_properties.csv', index_col=0)
Nb_properties_df = pd.read_csv('Output/Nb_compounds_properties.csv', index_col=0)
Ni_properties_df = pd.read_csv('Output/Ni_compounds_properties.csv', index_col=0)
Ta_properties_df = pd.read_csv('Output/Ta_compounds_properties.csv', index_col=0)
Ti_properties_df = pd.read_csv('Output/Ti_compounds_properties.csv', index_col=0)
V_properties_df = pd.read_csv('Output/V_compounds_properties.csv', index_col=0)
W_properties_df = pd.read_csv('Output/W_compounds_properties.csv', index_col=0)
Zr_properties_df = pd.read_csv('Output/Zr_compounds_properties.csv', index_col=0)






```

```python
# Append materials for merged properties dataframe
merged_properties_df = C_properties_df.append([Cr_properties_df, Fe_properties_df, Hf_properties_df, Mo_properties_df, Nb_properties_df, Ni_properties_df, Ta_properties_df, Ti_properties_df, V_properties_df, W_properties_df, Zr_properties_df])
merged_properties_df
```

```python
# Append materials for merged composition dataframe
merged_composition_df = C_compounds_df.append([Cr_compounds_df, Fe_compounds_df, Hf_compounds_df, Mo_compounds_df, Nb_compounds_df, Ni_compounds_df, Ta_compounds_df, Ti_compounds_df, V_compounds_df, W_compounds_df, Zr_compounds_df])
merged_composition_df
```

```python
# Merge composition and properties dataset
total_dataset_df = pd.merge(merged_properties_df, merged_composition_df, how='inner', left_on = ['mp_id'], right_on = ['mp_id'])
total_dataset_df
```

```python
# write to csv for database integration
merged_composition_df.to_csv(r'Output/merged_composition.csv', index = True)
merged_properties_df.to_csv(r'Output/merged_properties.csv', index = True)
total_dataset_df.to_csv(r'Output/total_dataset.csv', index = True)
```
